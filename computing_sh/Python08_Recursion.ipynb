# **Recursion**


* a divide and conquer problem solving methodology 
* program calls itself with (a) smaller and similar subproblem(s) (recursive case(s))
* until its base/anchor/terminating case(s) which can be solved directly
* program then backtracks to solve the problem

```
# Factorial
# eg factorial(5) = 5 * 4 * 3 * 2 * 1
#                 = 5 * factorial(4)
#                 = 5 * 4 * factorial(3)
#                 = 5 * 4 * 3 * factorial(2)
#                 = 5 * 4 * 3 * 2 * factorial(1)
#                 = 5 * 4 * 3 * 2 * 1
# factorial(n) = n * factorial(n-1) - recursive case
# factorial(1) = 1 - terminating case
# factorial(0) = 1 - terminating case

# using loop / iteration / repetition
def factoriali(n):
  result = 1
  for i in range(n, 0, -1):
    result = result * i
  return result

# using recursion
def factorialr(n):
  #if n > 0: # recursive case
  #  return n * factorialr(n-1)
  #else: # terminating case
  #  return 1
  if n == 0: # terminating case
    return 1
  else: # recursive case
    return n * factorialr(n-1) # tail recursion

# main
print(factoriali(1000))
print(factorialr(1000))
```

```
# Fibonacci: 1, 1, 2, 3, 5, 8, 13, 21
# fib(n) = fib(n-1) + fib(n-2) - recursive cases
# fib(1) = 1, fib(2) = 1 - terminating cases

# using recursion
def fibr(n):
  if n <= 2: # (n == 1) or (n == 2)
    return 1
  else:
    return fibr(n-1) + fibr(n-2)

# using loop
def fibi(n):
  if n <= 2:
    return 1
  else:
    fn1, fn2 = 1, 1
    result = 2
    for i in range(3, n+1):
      result = fn1 + fn2
      fn1 = fn2
      fn2 = result
    return result

# using dynamic programming
def fib_dp(n):
  memo = [1, 1]
  for i in range(1, n-1):
    memo.append(memo[i]+memo[i-1])
  return memo[n-1]

# main
print(fib_dp(50))
print(fibi(50))
print(fibr(50))
```
